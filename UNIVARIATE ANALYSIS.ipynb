```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate
# Đọc tệp dữ liệu
data = pd.read_csv('user_behavior_dataset.csv')

# 1. Tạo bảng thống kê mô tả
summary_stats = data.describe()

# Hiển thị bảng thống kê mô tả dưới dạng bảng đẹp trong terminal
print("Thống kê mô tả:")
print(tabulate(summary_stats, headers='keys', tablefmt='fancy_grid'))

# 2. Vẽ hình dạng phân phối mỗi biến
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(data[column], kde=True, bins=30, color='blue')
    plt.title(f"Phân phối của biến {column}")
    plt.xlabel(column)
    plt.ylabel("Tần suất")
    plt.show()

# 3. Vẽ biểu đồ boxplot
for column in data.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 5))
    sns.boxplot(x=data[column], color='cyan')
    plt.title(f"Boxplot của biến {column}")
    plt.xlabel(column)
    plt.show()
```
